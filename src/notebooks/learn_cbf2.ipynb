```python
from collections import Counter
```

```python
def compute_tf(self, text):
        words = text.split()
        freq = Counter(words)

        # cari max frequency
        if len(freq) == 0:
            max_freq = 1
        else:
            max_freq = max(freq.values())

        tf = {}

        for word in freq:
            count = freq[word]
            tf[word] = count / max_freq

        return tf
```
