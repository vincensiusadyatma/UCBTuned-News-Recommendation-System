```python


import pandas as pd
import numpy as np

df_cosine = pd.read_csv("../../dataset/cosine_similarity.csv", index_col=0)  # Judul sebagai index
print("Data cosine similarity:")
print(df_cosine.head())

articles = df_cosine.columns.tolist()
K = len(articles)
print(f"\nJumlah artikel (arms): {K}")

np.random.seed(42) 
N_per_article = 10

rows = []
for article in articles:
    for i in range(1, N_per_article+1):
        reward = np.random.binomial(1, 0.5)  
        rows.append({
            'Artikel': article,
            'Iterasi': i,
            'Reward': reward
        })

reward_history_df = pd.DataFrame(rows)

print("\nReward history dummy (10 interaksi per artikel):")
print(reward_history_df.head(20))  

reward_history_df.to_csv("../../dataset/reward_history_dummy.csv", index=False)
print("\nReward history berhasil disimpan ke '../../dataset/reward_history_dummy.csv'")
```
