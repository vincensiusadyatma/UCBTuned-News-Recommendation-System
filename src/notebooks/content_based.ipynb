```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("../../dataset/news_preprocess.csv")
df
```

```python
tfidf = TfidfVectorizer() 
tfidf_matrix = tfidf.fit_transform(df['processed_text'])
```

```python
feature_names = tfidf.get_feature_names_out()

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), index=df['Judul'], columns=feature_names)

tfidf_df.to_csv('../../dataset/tfidf_matrix.csv')
```

```python
df = pd.read_csv("tfidf_matrix.csv")
df
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python

indices = pd.Series(df.index, index=df['Judul']).drop_duplicates()
```

```python
cosine_df = pd.DataFrame(cosine_sim, index=df['Judul'], columns=df['Judul'])

cosine_df.to_csv('../../dataset/cosine_similarity.csv')
```
